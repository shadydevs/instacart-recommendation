```python
aisles = pd.read_csv('data/aisles.csv')
departments = pd.read_csv('data/departments.csv')
order_products_prior = pd.read_csv('data/order_products__prior.csv')
order_products_train = pd.read_csv('data/order_products__train.csv')
orders = pd.read_csv('data/orders.csv')
products = pd.read_csv('data/products.csv')
```

```python
files = [aisles, departments, order_products_prior, order_products_train, orders, products]
```

```python
for file in files:
    file.head()
    file.info()
```

```python
# products_orders = pd.read_csv('products_all.csv')
products_orders = order_products_prior.append(order_products_train)
```

```python
products_orders
```

```python
products_orders = products_orders[['order_id', 'product_id', 'add_to_cart_order']]
```

### Combining files

```python
Full_PO = pd.merge(products_orders, orders, on='order_id')
```

```python
Full_PO = pd.merge(Full_PO, products, on='product_id')
```

```python
Full_PO
```

```python
data = Full_PO[['order_id','department_id', 'add_to_cart_order', 'order_number', 'order_dow', 'order_hour_of_day', 'days_since_prior_order']]
```

## Preprocessing

```python
data = pd.merge(data, departments, on='department_id')
```

```python
data = data.drop(['department_id'], axis=1)
```

```python
dataDropped = data.dropna()
```

```python
dataDropped
```

```python
sortedData = dataDropped.sort_values(['order_id', 'add_to_cart_order'], ascending=[True, False], ignore_index=True)
finalDf = sortedData.groupby('order_id').head(1)
departmentData = dataDropped[['order_id','department']]
departmentData = pd.get_dummies(departmentData, columns=['department'])
final = pd.merge(finalDf, departmentData.groupby('order_id').sum(), on='order_id')
```

### Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('instacart.csv')
```

```python
df.columns
```

```python
df = df.drop(['order_id'], axis=1)
```

```python
df = df.rename(columns={'add_to_cart_order': 'cart_size', 'order_number': 'previous_orders', 'department': 'lod'})
```

```python
df
```

```python
df['cart_size']-=1
```

Using a random sample of 10% of the data

```python
sample = df.sample(300000, random_state=1)
```

```python
X = sample.drop('lod', axis=1)
y = sample['lod']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=10)
```

```python
x_train.shape
y_train.shape
```

```python
y_train
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
prediction = knn.predict(x_test)
```

```python
# prediction
```

```python
# confusion_matrix(y_test,prediction)
# pd.crosstab(y_test, prediction, rownames = ['Actual'], colnames =['Predicted'], margins = True)
print('knn precision: ', precision_score(y_test, prediction, average='weighted'))
print('knn recall: ', recall_score(y_test, prediction, average='weighted'))
```

```python
dt = DecisionTreeClassifier(criterion = "entropy", random_state = 100, max_depth = 10, min_samples_leaf = 5)
```

```python
dt.fit(x_train, y_train)
```

```python
dtPrediction = dt.predict(x_test)
```

```python
print('decision precision: ', precision_score(y_test, dtPrediction, average='weighted'))
print('decision recall: ', recall_score(y_test, dtPrediction, average='weighted'))
```
